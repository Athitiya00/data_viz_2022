## ต่อตารางในแนวแกน X (เพิ่มรายละเอียดของข้อมูล)

```python
personal_data = data_df[['No.','sex','age','nationality']]
personal_data.head()
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']]
covid_data.head()
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1)
covid_data.head() #sample(frac=1) คำสั่งสลับแถว
```

### เอาตารางสองอันมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data)
personalcovid_data.head()
```

```python
personalcovid_data = covid_data.merge(personal_data)
personalcovid_data.head()
```

### List(row oriented)

```python
data = [ ['Jomes LLC',150,200,140],
             ['Alpha Co',200,210,215],
             ['Blue Inc',50,90,95] ]
header = ['account','Jan','Feb','Mar']
```

```python
df2 = pd.DataFrame.from_records(data,columns=header)
df2
```
